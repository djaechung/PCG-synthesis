# Objective:
Create a mapping of spectrogram IDs to murmur labels for all ai-generated spectrograms.

# Setup

The AI_generated spectrograms are named such that that their filename includes their label. We just need a function to extract that label from each filename.

```python
def label_from_filename(filename):
  lowercase_label = filename.split('_')[1]
  return f'{lowercase_label[0].upper()}{lowercase_label[1:]}'
```

Noting where our ai-generated spectrograms are stored

```python
SPEC_DATA_DIR = '/content/drive/MyDrive/Stuff I Coded/PCG_synthesis/data2/ai/spectrograms/'
SPEC_TRAIN_DIR = SPEC_DATA_DIR + 'train/'
SPEC_VAL_DIR = SPEC_DATA_DIR + 'val/'
SPEC_TEST_DIR = SPEC_DATA_DIR + 'test/'
```

# Making the Key

```python
ai_murmur_key = pd.DataFrame()
```

```python
filenames = []
labels = []
```

```python
# for the train, val, and test directories, record all spectrogram
# file names and their corresponding labels
for dir in [SPEC_TRAIN_DIR, SPEC_VAL_DIR, SPEC_TEST_DIR]:
  for filename in os.listdir(dir):
    filename = filename[:-4]
    filenames.append(filename)
    label = label_from_filename(filename)
    labels.append(label)
```

```python
ai_murmur_key['pcg_id'] = filenames
ai_murmur_key['murmur'] = labels
```
