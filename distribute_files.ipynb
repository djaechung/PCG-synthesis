# File Distribution

````python
data_dir = "/content/drive/MyDrive/Stuff I Coded/PCG_synthesis/data"
````

````python
# this csv file contains each wav file's name and associated split.
# we need this information to assign wav files to their respective split folders
data_key = pd.read_csv(f"{data_dir}/circor_digiscope_by_pcg.csv")
````

````python
# take all wav files from the 'raw' directory and move them to the train, val,
# test directories according to the splits specified in data_key
for i, row in data_key.iterrows():
    pcg_id = row["pcg_id"]
    src_path = f"{data_dir}/raw/{pcg_id}.wav"
    split = row["split"]
    dest_dir = f"{data_dir}/{split.lower()}"
    dst_path = f"{dest_dir}/{pcg_id}.wav"
    shutil.move(src_path, dst_path)
````
